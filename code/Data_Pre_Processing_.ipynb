```python
data_processed = data_processed.drop(axis=0,columns=['ca','thal'])
```

Attribute "ca" has 607 missing values in 617 values, any replacement of the missing values will not bring any significant use for later analsis, so we delete this attribute.
So is the case for attributes "thal", they have missing values for over 50% of the total samples.

```python
data_processed = data_processed.dropna(axis=0, how='any',subset=['restecg'])
data_processed
```

```python
data_processed['trestbps'] = pd.to_numeric(data_processed['trestbps'])
data_processed['trestbps']

trestbps_mean = data_processed['trestbps'].mean()
trestbps_mean = np.round(trestbps_mean,1)
trestbps_mean
data_processed['trestbps']=data_processed['trestbps'].fillna(value=trestbps_mean)
```

```python
data_processed['thalach']=pd.to_numeric(data_processed['thalach'])

thalach_mean = data_processed['thalach'].mean()
thalach_mean = np.round(thalach_mean,1)
thalach_mean
data_processed['thalach']=data_processed['thalach'].fillna(value=thalach_mean)
```
